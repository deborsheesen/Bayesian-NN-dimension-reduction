```python
import matplotlib.pyplot as plt, numpy as np, numpy.random as npr, csv
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import LocallyLinearEmbedding
from tqdm import trange
#%matplotlib notebook
%matplotlib inline
import torch.nn as nn, timeit, torch, pystan as ps
from time import time
import matplotlib.pyplot as plt
from pylab import plot, show, legend
```

```python
sm_constrainedX = ps.StanModel(file="nn_constrainedX_tanh.stan")
sm_constrainedtheta = ps.StanModel(file="nn_constrainedtheta_tanh.stan")
```

```python
M = 1000
X = 10*npr.rand(2,M)
Y = np.zeros((3,M))
Y[0] = np.cos(X[0])*np.cos(X[1])
Y[1] = np.cos(X[0])*np.sin(X[1])
Y[2] = np.sin(X[0])

idx = Y[2]>0
Y = Y[:,idx]
Y += 0.1*npr.rand(*np.shape(Y))
```

```python
data = dict(Nobs=np.shape(Y)[-1], in_dim=2, out_dim=np.shape(Y)[0], hidden_dim=50, y=Y)
```

#### Run Stan:

```python
control = dict(adapt_delta=0.9, max_treedepth=10)
start = time()
fit_constrainedX = sm_constrainedX.sampling(data=data, thin=1, n_jobs=4, chains=4,
                                            init="random", iter=500, control=control)
fit_constrainedtheta = sm_constrainedtheta.sampling(data=data, thin=1, n_jobs=4, chains=4,
                                            init="random", iter=500, control=control)
print(round((time()-start)/60,2), "minutes to run")
```

```python
trace_constrainedX = fit_constrainedX.extract()
trace_constrainedtheta = fit_constrainedtheta.extract()
n_samples = np.shape(trace_constrainedX ['X'])[0]
n_latent = np.shape(trace_constrainedX ['X'])[1]
```

### Traceplots of some latent variables:

#### Constrained $X$:

```python
plt.figure(figsize=(20,15))
for i in range(10):
    for j in range(10):
        k = 10*i+j
        plt.subplot(10,10,k+1)
        plt.plot(trace_constrainedX['X'][:,k,0])
        if j > 0 : 
            plt.yticks([])
plt.show()
```

#### Constrained $\theta$:

```python
plt.figure(figsize=(20,15))
for i in range(10):
    for j in range(10):
        k = 10*i+j
        plt.subplot(10,10,k+1)
        plt.plot(trace_constrainedtheta['X'][:,k,0])
        if j > 0 : 
            plt.yticks([])
plt.show()
```

### Loss:

```python
def L2_loss(Y, X, weights_1, bias_1, weights_2, bias_2) :
    Y_pred = np.matmul(np.tanh(np.matmul(X,weights_1)+bias_1), weights_2) + bias_2
    return np.mean(np.sum((Y-Y_pred)**2,axis=1))
```

```python
loss_constrainedX = np.zeros(n_samples)
loss_constrainedtheta = np.zeros(n_samples)

for i in trange(n_samples) :
    loss_constrainedX[i] = L2_loss(Y.transpose(), trace_constrainedX['X'][i], 
                                   trace_constrainedX['weights_1'][i], 
                                   trace_constrainedX['bias_1'][i], 
                                   trace_constrainedX['weights_2'][i], 
                                   trace_constrainedX['bias_2'][i]) 
                #/ trace_constrainedX['error_sigma2'][i]
    loss_constrainedtheta[i] = L2_loss(Y.transpose(), trace_constrainedtheta['X'][i], 
                               trace_constrainedtheta['weights_1'][i], 
                               trace_constrainedtheta['bias_1'][i], 
                               trace_constrainedtheta['weights_2'][i], 
                               trace_constrainedtheta['bias_2'][i]) 
plt.figure(figsize=(10,3))
plt.subplot(121)
plt.plot(loss_constrainedX)
plt.title(r"Constrained $X$")
plt.subplot(122)
plt.plot(loss_constrainedtheta)
plt.title(r"Constrained $\theta$")
plt.suptitle("Loss");
```

### Pairwise distances:

```python
idx = npr.choice(np.shape(trace_constrainedX['X'])[1],10,replace=False)
pw_dist_matrix_constrainedX = np.zeros((n_latent,n_latent,n_samples))
pw_dist_matrix_constrainedtheta = np.zeros((n_latent,n_latent,n_samples))

for i in trange(10) :
    for j in range(10) :
        for k in range(n_samples) :
            pw_dist_matrix_constrainedX[i,j,k] = np.linalg.norm(trace_constrainedX['X'][k,idx[i]]-
                                                       trace_constrainedX['X'][k,idx[j]])
            pw_dist_matrix_constrainedtheta[i,j,k] = np.linalg.norm(trace_constrainedtheta['X'][k,idx[i]]-
                                                       trace_constrainedtheta['X'][k,idx[j]])
```

#### Constrained $X$:

```python
plt.figure(figsize=(20,15))
for i in range(10):
    for j in range(10):
        if i > j :
            plt.subplot(10,10,10*i+j+1)
            plt.plot(pw_dist_matrix_constrainedX[i,j])
            plt.xticks([])
            if j > 0 : plt.yticks([])
plt.show()
```

```python
plt.figure(figsize=(20,15))
for i in range(10):
    for j in range(10):
        if i > j :
            plt.subplot(10,10,10*i+j+1)
            plt.plot(pw_dist_matrix_constrainedtheta[i,j])
            plt.xticks([])
            if j > 0 : plt.yticks([])
plt.show()
```
