```python
import torch, torch.nn as nn, copy, timeit, numpy.random as npr, numpy as np
from torch.distributions.bernoulli import Bernoulli 
from time import time
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import plot, show, legend
from scipy.stats import uniform
```

#### Define number of layers and number of nodes per layer:

```python
n_in = 1
n_h1 = 200
n_h2 = 20
n_out = 2
```

#### Define model:

```python
model = nn.Sequential(nn.Linear(n_in, n_h1),
                      nn.Tanh(),
                      nn.Linear(n_h1, n_h2),
                      nn.Tanh(),
                      nn.Linear(n_h2, n_out))
```

```python
print(model)
```

#### Simulate some data:

```python
error_sigma2 = 0.1
Nobs = 2_000
X = torch.randn(Nobs, n_in, requires_grad=True)
X = X.data
Y = np.zeros((Nobs, n_out))
Y[:,0] = list(abs(X)*np.cos(2*X))
Y[:,1] = list(abs(X)*np.sin(3*X))
Y = 5*torch.from_numpy(Y).float() + torch.randn(Nobs,2)*np.sqrt(error_sigma2)
criterion = nn.MSELoss()
plt.plot(list(Y[:,0]), list(Y[:,1]), 'o', markersize=2)
plt.grid(True)
plt.title("True y");
```

#### Randomly initialise model weights:

```python
def init_normal(m):
    if type(m) == nn.Linear:
        nn.init.normal_(m.weight)

def init_uniform(m):
    if type(m) == nn.Linear:
        nn.init.uniform_(m.weight)
```

```python
model.apply(init_normal)
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3, momentum=0.9, nesterov=True)
```

### NN training with randomly perturbed input:

```python
X_perturbed = X + 1e-1*torch.randn(np.shape(X))
```

```python
n_iter = 50_000
start = time()
for epoch in range(n_iter):
    y_pred = model(X_perturbed)
    loss = criterion(y_pred, Y)
    if (epoch+1)%(n_iter/10) == 0 :
        print("Epoch: %d; loss: %.3f; time: %.1f mins" % (epoch+1, loss.item(), (time()-start)/60))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plt.figure(figsize=(5,5))
plt.plot(list(Y[:,0]),list(Y[:,1]), 'ro', label="observed", markersize=2)
plt.plot(list(model(X_perturbed)[:,0]),list(model(X_perturbed)[:,1]), 'bo', label="predicted", markersize=2)
plt.legend()
plt.grid(True)
```
